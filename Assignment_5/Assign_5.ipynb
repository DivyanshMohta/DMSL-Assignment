```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Importing Dataset
data = pd.read_csv("/content/car_crashes.csv")
df = pd.DataFrame(data)
df.head()
```

```python
# BAR GRAPH
df.plot(kind='bar',
        x='alcohol',
        y='not_distracted',
        color='orange',
        width=0.3)

plt.title('BarPlot')
plt.show()
```

```python
# LINE GRAPH
plt.plot(data['total'],data['speeding'], 'g')
plt.xlabel("Total")
plt.ylabel("Speeding")
plt.title("Line Chart")
plt.show()
```

```python
# SCATTER PLOT
plt.scatter(data['abbrev'],data['no_previous'],color='m',s=70,edgecolor='black')
plt.xlabel("X-axis",fontsize=11)
plt.ylabel("Y-axis",fontsize=11)
plt.title("Scatter Plot")
plt.show()
```

```python
# HISTOGRAM
df['total'].plot(kind='hist',color='yellow', bins=20, edgecolor='black',width=0.7,linewidth=1.5)
plt.title('Histogram Plot')
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.violinplot(x=data['total'], y=data['speeding'],)
plt.title('Violin Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
# BOX PLOT
df.plot.box()
plt.title("BOX PLOT")
plt.show()
```

```python
# AREA PLOT
plt.fill_between(data['abbrev'],data['total'])
plt.title("AREA PLOT")
plt.xlabel("Abbrev")
plt.ylabel("total")
# plt.figure(figsize=(20,20))
# plt.figure().set_figwidth(50)

plt.show()
```

```python
# STAIRS PLOT
plt.stairs(data['speeding'], linewidth=2.5)
plt.title("STAIRS PLOT")
plt.show()
```
